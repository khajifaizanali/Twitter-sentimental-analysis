```python
tweet=[]
ftest=[]
for i in tweets:
    a=" ".join(i)
    tweet.append(a)
for i in ttest:
    a=" ".join(i)
    ftest.append(a)
from sklearn.feature_extraction.text import CountVectorizer
count_vec=CountVectorizer(max_features=8000,stop_words=stop,max_df=0.8,ngram_range =(1,3))
tweet
```

```python
a=count_vec.fit_transform(tweet)
b=count_vec.transform(ftest)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
clf = SVC()
grid = {'C' : [1e2, 1e3, 5e3, 1e4, 5e4, 1e5],
       'gamma' : [1e-3, 5e-4, 1e-4, 5e-3]}
abc = GridSearchCV(clf, grid)
abc.fit(a, y)
```
